```python
# Mering by a common key

df1 = pd.DataFrame({
    'key1':['K0','K8','K2','K3'],
    'A':['A0','A1','A2','A3'],
    'C':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3']
})

df2 = pd.DataFrame({
    'key2':['K0','K1','K2','K3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
})
```

```python
print(df1)
print()
print(df2)
```

```python
df1
```

```python
df2
```

```python
df3 = pd.merge(df1,df2,how='left',left_on='key1',right_on='key2')
df3
```

```python
df4 = pd.merge(df1,df2,how='right',left_on='key1',right_on='key2')
df4
```

```python
left = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                    'key2':['K0','K1','K2','K3'],
                    'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']
                    })

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']}
                    )
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on=['key1','key2'])
```

```python
pd.merge(left,right,how='left',on=['key1','key2'])
```

```python
pd.merge(left,right,how='right',on=['key1','key2'])
```

```python
# Join Operator

left1 = pd.DataFrame({'A':['A0','A1','A2'],
                     'B':['B0','B1','B2']},
                    index=['K0','K1','K2'])

right1 = pd.DataFrame({'C':['C0','C1','C2'],
                      'D':['D0','D1','D2']},
                     index=['K0','K2','K3'])
```
