```python
# Multi-Indexing

outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]

heir_index = list(zip(outside,inside))
print(heir_index)
print(type(heir_index))
```

```python
heir_index1 = zip(outside,inside)
print(heir_index1)
print(type(heir_index1))
```

```python
heir_index1 = tuple(zip(outside,inside))
print(heir_index1)
print(type(heir_index1))
```

```python
heir_index = pd.MultiIndex.from_tuples(heir_index)
print(heir_index)
```

```python
df7 = pd.DataFrame(data=np.round(rn(6,3)),index=heir_index,columns=['A','B','C'])
df7
```

```python
print(df7.xs('G2'))
```

```python
print(df7.xs('G1'))
```
